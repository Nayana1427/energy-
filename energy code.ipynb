```python
df['Date_Time'] = pd.to_datetime(df['Date_Time'], errors='coerce')
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna({
    'Charging_Load_kW': df['Charging_Load_kW'].median(),
    'Energy_Drawn_kWh': df['Energy_Drawn_kWh'].median(),
    'Queue_Time_mins': df['Queue_Time_mins'].median(),
    'Temperature_C': df['Temperature_C'].mean(),
    'Wind_Speed_m/s': df['Wind_Speed_m/s'].mean(),
    'Precipitation_mm': 0,
}, inplace=True)
```

```python
df = df[
    (df['Charging_Load_kW'] >= 0) &
    (df['Energy_Drawn_kWh'] >= 0) &
    (df['State_of_Charge_%'].between(0, 100)) &
    (df['Battery_Capacity_kWh'] > 0)
]
```

```python
df['Road_Conditions'] = df['Road_Conditions'].str.strip().str.title()
df['Weather_Conditions'] = df['Weather_Conditions'].str.strip().str.title()
```

```python
df['Hour'] = df['Date_Time'].dt.hour
df['Day'] = df['Date_Time'].dt.day
df['Month'] = df['Date_Time'].dt.month
df['Year'] = df['Date_Time'].dt.year
df['Weekday_Name'] = df['Date_Time'].dt.day_name()
```

```python
df.sort_values(by='Date_Time', inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv("cleaned_ev_charging_dataset.csv", index=False)

print("✅ Data cleaned successfully!")
print(f"Rows after cleaning: {len(df)}")
print("Saved as 'cleaned_ev_charging_dataset.csv'")
```
